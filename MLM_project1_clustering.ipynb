K-Means Clustering 1 Determine Value of 'K' in K-Means using Elbow Curve & KMeans-Inertia

```python
from sklearn.cluster import KMeans as kmclus
import matplotlib.pyplot as plt

wcssd = []
nr_clus = range(1, 11)  # Number of Clusters
for k in nr_clus:
    kmeans = kmclus(n_clusters=k, init='random', random_state=111)
    kmeans.fit(survey_subset)
    wcssd.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.plot(nr_clus, wcssd, marker='x')
plt.xlabel('Values of K')
plt.ylabel('Within Cluster Sum Squared Distance')
plt.title('Elbow Curve for Optimal K')
plt.show()
```

```python
from scipy.stats import f_oneway

# Davies-Bouldin scores for k=2, k=3, k=4, and k=5
dbscore_k_2 = [0.5212779676911768]
dbscore_k_3 = [0.5249914766911825]
dbscore_k_4 = [0.5301161749621922]
dbscore_k_5 = [0.5399317340477215]

# Perform ANOVA
f_statistic, p_value = f_oneway(dbscore_k_2, dbscore_k_3, dbscore_k_4, dbscore_k_5)

# Print results
print("F-statistic:", f_statistic)
print("p-value:", p_value)

# Interpret results
if p_value < 0.05:
    print("Reject null hypothesis: There is a significant difference in DB scores for different values of k.")
else:
    print("Fail to reject null hypothesis: There is no significant difference in DB scores for different values of k.")
```
